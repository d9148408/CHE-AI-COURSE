### **I. 練習題: Pandas 資料處理與分析**
 - 使用下面化工製程模擬數據集進行練習。

````python
# 生成化工製程模擬數據集
np.random.seed(42)

# 生成時間序列 (連續24小時的製程數據, 每5分鐘記錄一次)
n_samples = 24 * 12  # 288筆數據
time_index = pd.date_range('2024-01-01 00:00', periods=n_samples, freq='5min')

# 生成製程數據
data = {
    '時間': time_index,
    '反應器溫度': 180 + 20 * np.sin(np.linspace(0, 4*np.pi, n_samples)) + np.random.normal(0, 3, n_samples),
    '反應器壓力': 10 + 2 * np.cos(np.linspace(0, 4*np.pi, n_samples)) + np.random.normal(0, 0.5, n_samples),
    '進料流量': 50 + 10 * np.sin(np.linspace(0, 2*np.pi, n_samples)) + np.random.normal(0, 2, n_samples),
    '產品濃度': 85 + 10 * np.cos(np.linspace(0, 3*np.pi, n_samples)) + np.random.normal(0, 3, n_samples),
    '冷卻水溫度': 25 + 5 * np.sin(np.linspace(0, 6*np.pi, n_samples)) + np.random.normal(0, 1, n_samples),
    '批次編號': ['Batch_' + str(i//48 + 1) for i in range(n_samples)]  # 每批次約4小時
}

# 建立 DataFrame
df = pd.DataFrame(data)

# 隨機插入一些缺失值 (模擬感測器故障)
missing_indices = np.random.choice(df.index, size=15, replace=False)
df.loc[missing_indices[0:5], '反應器溫度'] = np.nan
df.loc[missing_indices[5:10], '反應器壓力'] = np.nan
df.loc[missing_indices[10:15], '產品濃度'] = np.nan

print("化工製程數據集已生成")
print(f"數據形狀: {df.shape}")
print(f"\n前5筆數據:")
print(df.head())
````

### **實驗4: 進階時間序列分析**

**任務描述**：
使用 Pandas 進階時間序列功能：
- 計算時間差 (time delta) 與製程事件持續時間
- 使用 `shift()` 計算變化率 (rate of change)
- 偵測製程異常 (例如: 溫度突然上升或下降)
- 計算移動平均交叉點 (用於判斷趨勢轉變)
- 分析製程週期性 (是否有規律的波動模式)

**提示**：
```python
# 計算變化率
df['temp_change'] = df['反應器溫度'].diff()  # 與前一筆的差值
df['temp_pct_change'] = df['反應器溫度'].pct_change()  # 百分比變化

# 偵測異常 (變化率超過閾值)
threshold = df['temp_change'].std() * 2
df['is_anomaly'] = abs(df['temp_change']) > threshold
```
